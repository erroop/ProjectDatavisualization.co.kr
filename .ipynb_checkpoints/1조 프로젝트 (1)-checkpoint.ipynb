```python
dong=['청담동', '신사동', '개포동','도곡동', '대치동', '역삼동', '논현동', '율현동', '압구정동', '삼성동', '세곡동', '수서동']
```

```python
count=0
for n in dong:
    count+=1
    if n == '개포동':
        d = pd.read_csv('./data/'+n+'.csv',encoding = 'cp949')
    elif n == '논현동':
        d = pd.read_csv('./data/'+n+'.csv',encoding = 'cp949')
    else:
        d = pd.read_csv('./data/'+n+'.csv')
    d ['Unnamed: 0'] = n
    d .rename(columns={'Unnamed: 0':'강남구'}, inplace = True)
    L.append(d)
    bdf=pd.concat(L)
print(count)
```

```python
# 강남구 동별 월평균매출 - 월세 최대값
max=pd.DataFrame(((df1.groupby('구동')['분기당매출'].sum())/(df1.groupby('구동')['점포수'].sum())/(df1['구동'].value_counts())*3) -(bdf.groupby('강남구')['월세'].max()))
max
```

```python
# 강남구 동별 월평균매출 - 월세 최소값
min=pd.DataFrame(((df1.groupby('구동')['분기당매출'].sum())/(df1.groupby('구동')['점포수'].sum())/(df1['구동'].value_counts())*3)-(bdf.groupby('강남구')['월세'].min()))
min
```

```python
# 강남구 동별 월평균매출 - 월세 평균값
mean=pd.DataFrame(((df1.groupby('구동')['분기당매출'].sum())/(df1.groupby('구동')['점포수'].sum())/(df1['구동'].value_counts())*3) -(bdf.groupby('강남구')['월세'].mean()))
mean
```

```python
condf = pd.concat([min, max, mean], axis=1)
condf
```

```python
# 각 동의 월평균매출에 월세최소값, 월세최대값, 월세평균값을 뺀 데이터
condf.columns =['최소', '최대', '평균']
condf
```
